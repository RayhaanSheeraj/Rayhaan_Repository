    <h1>The Mysterious Story of the Flailing Woods</h1>
    <div id="gameGrid"></div>
    <script>
        const grid = document.getElementById("gameGrid");
        const gridSize = 4;
        let playerPos = { x: 0, y: 0 };
        let npcPos = { x: 3, y: 3 };
        function initializeGrid() {
            for (let row = 0; row < gridSize; row++) {
                for (let col = 0; col < gridSize; col++) {
                    const cell = document.createElement("div");
                    cell.classList.add("grid-cell");
                    if (row === playerPos.y && col === playerPos.x) {
                        cell.textContent = "X";
                    } else if (row === npcPos.y && col === npcPos.x) {
                        cell.textContent = "NPC";
                    }
                    grid.appendChild(cell);
                }
            }
        }
        function updateGrid() {
            const cells = document.querySelectorAll(".grid-cell");
            cells.forEach(cell => {
                cell.textContent = "";
            });
            cells[playerPos.y * gridSize + playerPos.x].textContent = "X";
            cells[npcPos.y * gridSize + npcPos.x].textContent = "NPC";
        }
        function checkCollision() {
            if (playerPos.x === npcPos.x && playerPos.y === npcPos.y) {
                const question = "Hello there! How can I help you?\nChoose a question:\n1. What is your name?\n2. What is this place?\n3. Where is everyone?";
                const answer = prompt(question);
                const responses = {
                    "1": "NPC: My name is Egalt.",
                    "2": "NPC: It's the woods, north from Frake Town. A lot of outcasts live in these cabins right here.",
                    "3": "NPC: Everyone slowly started disappearing. Something isn't right..."
                };
                if (responses[answer]) {
                    alert(responses[answer]);
                } else {
                    alert("Please choose a valid option (1, 2, or 3).");
                }
            }
        }
        document.addEventListener('keydown', (e) => {
            switch (e.key) {
                case 'w':
                case 'W':
                    if (playerPos.y > 0) playerPos.y--;
                    break;
                case 's':
                case 'S':
                    if (playerPos.y < gridSize - 1) playerPos.y++;
                    break;
                case 'a':
                case 'A':
                    if (playerPos.x > 0) playerPos.x--;
                    break;
                case 'd':
                case 'D':
                    if (playerPos.x < gridSize - 1) playerPos.x++;
                    break;
            }
            updateGrid();
            checkCollision();
        });
        initializeGrid();
    </script>
</body>

<body>
    <h2>Sprite Sheet Animation</h2>
    <canvas id="spriteCanvas"></canvas>
    <script>
        let img = new Image();
        let currentFrame = 0;
        let frameCount = 0;
        let ctx, canvas;
        const spriteUrl = 'https://static.wixstatic.com/media/c6fb96_88ba297597bc4ad0812e06be20cb28c9~mv2.png/v1/fill/w_1600,h_900,al_c,q_90/file.jpg';
        const frameWidth = 64;
        const frameHeight = 64;
        const totalFrames = 16;
        const frameSpeed = 10;
        function startAnimation() {
            canvas = document.getElementById("spriteCanvas");
            canvas.width = frameWidth;
            canvas.height = frameHeight;
            ctx = canvas.getContext("2d");
            img.src = spriteUrl;
            img.onload = displaySprite;
        }
        function displaySprite() {
            const col = currentFrame % (totalFrames / 4);
            const row = Math.floor(currentFrame / (totalFrames / 4));
            ctx.clearRect(0, 0, canvas.width, canvas.height);
            ctx.drawImage(img, col * frameWidth, row * frameHeight, frameWidth, frameHeight, 0, 0, frameWidth, frameHeight);
            frameCount++;
            if (frameCount >= frameSpeed) {
                currentFrame = (currentFrame + 1) % totalFrames;
                frameCount = 0;
            }
            requestAnimationFrame(displaySprite);
        }
        window.onload = startAnimation;
    </script>
</body>
