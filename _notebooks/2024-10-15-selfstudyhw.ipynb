    <div class="game-map" id="gameMap"></div>
    <button class="button" onclick="interactWithNpc()">Interact with NPC</button>
    <script>
        const gameMap = [
            [" ", " ", " ", " ", " "],
            [" ", "NPC", " ", " ", " "],
            [" ", " ", " ", " ", " "],
            [" ", " ", " ", " ", " "],
            [" ", " ", " ", " ", " "]
        ];
        function displayGameMap(map) {
            const gameMapElement = document.getElementById('gameMap');
            gameMapElement.innerHTML = ''; // Clear previous map
            map.forEach(row => {
                row.forEach(cell => {
                    const cellElement = document.createElement('div');
                    cellElement.classList.add('cell');
                    if (cell === 'NPC') {
                        cellElement.classList.add('npc');
                        cellElement.textContent = 'NPC';
                    }
                    gameMapElement.appendChild(cellElement);
                });
            });
        }
        function interactWithNpc() {
            const playerPosition = { x: 1, y: 1 }; // Simulate player position
            if (gameMap[playerPosition.x][playerPosition.y] === 'NPC') {
                const questions = [
                    "1. What is your name?",
                    "2. What is your quest?",
                    "3. What is your favorite color?"
                ];
                let questionText = "NPC: Hello, traveler! Do you have any questions?\n";
                questions.forEach(question => {
                    questionText += question + "\n";
                });
                const answer = prompt(questionText + "Choose a question to ask (1-3):");
                let response = "NPC: I don't understand your question.";
                if (answer === "1") {
                    response = "NPC: My name is Eldor.";
                } else if (answer === "2") {
                    response = "NPC: I seek the Holy Grail.";
                } else if (answer === "3") {
                    response = "NPC: Blue, no, wait, yellow!";
                }
                alert(response);
            } else {
                alert("There is no one here to interact with.");
            }
        }
        displayGameMap(gameMap);
    </script>
