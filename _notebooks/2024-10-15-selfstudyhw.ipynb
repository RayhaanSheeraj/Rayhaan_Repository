    <title>NPC-Interation Game Hack</title>
    <style>
        body {
            font-family: Arial, sans-serif;
        }
        #gameGrid {
            display: grid;
            grid-template-columns: repeat(4, 50px);
            grid-template-rows: repeat(4, 50px);
            gap: 5px;
            margin: 20px;
        }
        .grid-cell {
            width: 50px;
            height: 50px;
            background-color: lightgray;
            display: flex;
            justify-content: center;
            align-items: center;
            font-size: 20px;
        }
    </style>

<body>
    <h1>The Mysterious Story of the Flailing Woods</h1>
    <div id="gameGrid"></div>
    <script>
        const grid = document.getElementById("gameGrid");
        const gridSize = 4;
        let playerPos = { x: 0, y: 0 };
        let npcPos = { x: 3, y: 3 };
        function initializeGrid() {
            for (let row = 0; row < gridSize; row++) {
                for (let col = 0; col < gridSize; col++) {
                    const cell = document.createElement("div");
                    cell.classList.add("grid-cell");
                    if (row === playerPos.y && col === playerPos.x) {
                        cell.textContent = "X";
                    } else if (row === npcPos.y && col === npcPos.x) {
                        cell.textContent = "NPC";
                    }
                    grid.appendChild(cell);
                }
            }
        }
        function updateGrid() {
            const cells = document.querySelectorAll(".grid-cell");
            cells.forEach(cell => {
                cell.textContent = "";
            });
            cells[playerPos.y * gridSize + playerPos.x].textContent = "X";
            cells[npcPos.y * gridSize + npcPos.x].textContent = "NPC";
        }
        function checkCollision() {
            if (playerPos.x === npcPos.x && playerPos.y === npcPos.y) {
                const question = "Hello there! How can I help you?\nChoose a question:\n1. What is your name?\n2. What is this place?\n3. Where is everyone?";
                const answer = prompt(question);
                const responses = {
                    "1": "NPC: My name is Egalt.",
                    "2": "NPC: It's the woods, north from Frake Town. A lot of outcasts live in these cabins right here.",
                    "3": "NPC: Everyone slowly started disappearing. Something isn't right..."
                };
                if (responses[answer]) {
                    alert(responses[answer]);
                } else {
                    alert("Please choose a valid option (1, 2, or 3).");
                }
            }
        }
        document.addEventListener('keydown', (e) => {
            switch (e.key) {
                case 'w':
                case 'W':
                    if (playerPos.y > 0) playerPos.y--;
                    break;
                case 's':
                case 'S':
                    if (playerPos.y < gridSize - 1) playerPos.y++;
                    break;
                case 'a':
                case 'A':
                    if (playerPos.x > 0) playerPos.x--;
                    break;
                case 'd':
                case 'D':
                    if (playerPos.x < gridSize - 1) playerPos.x++;
                    break;
            }
            updateGrid();
            checkCollision();
        });
        initializeGrid();
    </script>
</body>
