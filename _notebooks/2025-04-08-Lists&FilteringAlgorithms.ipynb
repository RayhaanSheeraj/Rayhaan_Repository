### Homework Hack 3

```python
import pandas as pd

def filter_spotify_streams(csv):
    spotify_streams = pd.read_csv(csv)

    count = 0

    for streams in spotify_streams["Total Streams (Millions)"]:
        if streams >= 10:
            print(str(spotify_streams["Album"][count]), "has over a million streams. It has ", str(streams), "million streams")
        count += 1

filter_spotify_streams("Spotify_2024_Global_Streaming_Data.csv")
```
